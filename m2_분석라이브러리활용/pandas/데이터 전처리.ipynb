## [과제] 아래 가이드에 따라서 고객 구매데이터를 생성하고 아래사항을 수행하세요(관측치 10000개)

#### Q1. 고객 속성(id, 성별, 연령, 거주지역) 데이터와 거래 데이터를 통합한 데이터 프레임 df를 생성하세요.

거래건수
- id : 1 ~ 10000번 일련번호(거래건수)

고객별 속성 데이터
- cusno : 고객번호(1000)
- gender : 0,1 정수 난수 생성
- age : 0 ~ 80사이 정수 난수 생성
- region : 1 ~ 10 사이 정수 난수 생성

거래 데이터
- day : 요일
- date : 2020-01-01 ~ 2021-12-31 사이의 날짜 생성 후 거래 년도에 해당하는 날짜를 랜덤하게 적용
- product_20 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10 ~ 30
- product_21 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10 ~ 30
- price_20 : 1000 ~ 50000 사이 실수 난수 생성
- price_21 : 1000 ~ 50000 사이 실수 난수 생성
- quantity_20 : 1 ~ 100 사이 정수 난수 생성
- quantity_21 : 1 ~ 100 사이 정수 난수 생성
- time_20 : 01 ~ 24 사이 정수 난수 생성
- time_21 : 01 ~ 24 사이 정수 난수 생성
- amount_20 : price_20 * quantity_20
- amount_21 : price_21 * quantity_21
- sales : 20년 대비 21년구매금액이 증가면 1, 감소면 0

```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# 고객 속성 데이터 생성
import numpy as np
import pandas as pd
np.random.seed(0)
no = np.arange(1,10001)
id1 = pd.Series(no)
gender = np.random.randint(2,size=10000)
gender1 = pd.Series(gender)
age=np.random.randint(10,81,size=10000)
age1 = pd.Series(age)
region = np.random.randint(1,11,size=10000)
region1 = pd.Series(region)
df1 = pd.concat([id1,gender1,age1,region1],axis=1)
df1.rename(columns={0:'no',1:'gender',2:'age',3:'region'},inplace=True)
df1.head()
```

```python
data = np.zeros((10000,11))
df2 = pd.DataFrame(data,columns=['cusno','day','date','product_20','product_21','price_20','price_21','quantity_20','quantity_21','time_20','time_21'])
df2.head()
```

```python
# 거래 데이터 생성
np.random.seed(1)
df2['cusno'] = np.random.randint(1,10001,10000)
df2['day'] = np.random.choice(['MON','TUE','WED','THU','FRI','SAT','SUN'],10000)
df2['date'] = pd.date_range(start='2020-01-01',end='2021-12-31',periods=10000)
df2['product_20'] = np.random.randint(1,4,10000)
df2['product_21'] = np.random.randint(1,4,10000)
df2['price_20'] = np.random.randint(1000,50001,10000)
df2['price_21'] = np.random.randint(1000,50001,10000)
df2['quantity_20'] = np.random.randint(1,101,10000)
df2['quantity_21'] = np.random.randint(1,101,10000)
df2['time_20'] = np.random.randint(1,25,size=10000)
df2['time_21'] = np.random.randint(1,25,size=10000)
df2['year'] = df2.date.dt.year
df2['sales'] = np.where(df2.amount_21>df2.amount_20,1,0)
df2.head()
```

```python
# product_20 고가 중가 저가 설정
def prod_diff_20(product_20,price_20):
    if product_20 == 1:
        return price_20 * 20
    elif product_20 == 2:
        return price_20 * 10
    else:
        return price_20 * 5
df2['price_20'] = df2.apply(lambda x : prod_diff_20(x['product_20'],x['price_20']),axis=1)
df2.head()
```

```python
# product_21 고가 중가 저가 설정
def prod_diff_21(product_21,price_21):
    if product_21 == 1:
        return price_21 * 20
    elif product_21 == 2:
        return price_21 * 10
    else:
        return price_21 * 5
df2['price_21'] = df2.apply(lambda x : prod_diff_21(x['product_21'],x['price_21']),axis=1)
df2.head()
```

```python
# quantity_20 1~100 소량 대량 나누기(확실한 구분점을 찾기위해)
def quan_diff_20(product_20,quantity_20):
    if product_20 >0 and product_20 < 50:
        return quantity_20 * 10
    else:
        return quantity_20
df2['quantity_20'] = df2.apply(lambda x : quan_diff_20(x['product_20'],x['quantity_20']),axis=1)
df2.head()
```

```python
# quantity_21 소량 대량 나누기
def quan_diff_21(product_21,quantity_21):
    if product_21 >0 and product_21 < 50:
        return quantity_21 * 10
    else:
        return quantity_21
df2['quantity_21'] = df2.apply(lambda x : quan_diff_21(x['product_21'],x['quantity_21']),axis=1)
df2.head()
```

```python
# amount 설정
df2['amount_20'] = df2.quantity_20 * df2.price_20
df2['amount_21'] = df2.quantity_21 * df2.price_21
df2.head()
```
